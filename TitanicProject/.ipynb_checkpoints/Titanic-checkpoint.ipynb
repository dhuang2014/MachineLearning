```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import time
```

```python
train_data = pd.read_csv('train.csv')
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
# data
survive = train_data.Survived
ages = train_data.Age
gender = train_data.Sex
pclass = train_data.Pclass
sibsp = train_data.SibSp
parch = train_data.Parch
```

```python
# Age
# plt.figure(figsize=(8,6))
print('total counts: ' + str(len(ages)))
print('age exsit counts:' + str((ages > -1).sum()))
print('\n')
case = 1
if case == 1:
    %matplotlib notebook
    %matplotlib inline
elif case == 2:
    %matplotlib inline
    from IPython.display import set_matplotlib_formats
    set_matplotlib_formats('pdf')
elif case == 3:
    %matplotlib notebook
font = {'family' : 'times new roman',
        'weight' : 'regular',
        'size'   : 18}
matplotlib.rc('font', **font)
plt.figure()
plt.hist(ages[~np.isnan(ages)],range=[0.5,80.5],bins = 80)
plt.title('Age Distribution')
plt.xlabel('Ages')
plt.ylabel('Counts')
plt.figure()
plt.scatter(ages, survive, s=1)
plt.xlabel('Ages')
plt.ylabel('Survive or Death')
plt.title('Age vs Survive (survive: 1, death: 0)')
plt.yticks(np.arange(0, 2., 1.))
plt.show()
print('\n')
# Conclusion from age distribution: 
print('Age is not relevant to death and survial') 
```

```python
# Gender/Sex
counts = len(gender)
print('total counts: ' + str(counts))
femaleC = (gender == 'female').sum()
print('female counts: ' + str(femaleC))
femaleSC = ((gender == 'female') & (survive == 1)).sum()
print('female survive counts: ' + str(femaleSC))
femaleDC = ((gender == 'female') & (survive == 0)).sum()
print('female death counts: ' + str(femaleDC))
femaleSR = ((gender == 'female') & (survive == 1)).sum()/(gender == 'female').sum()
print('female survive ratio: ' + str(femaleSR))

maleC = (gender == 'male').sum()
print('male counts: ' + str(maleC))
maleSC = ((gender == 'male') & (survive == 1)).sum()
print('male survive counts: ' + str(maleSC))
maleDC = ((gender == 'male') & (survive == 0)).sum()
print('male death counts: ' + str(maleDC))
maleSR = ((gender == 'male') & (survive == 1)).sum()/(gender == 'male').sum()
print('male survive ratio: ' + str(maleSR))
print('\n')

case = 1
if case == 1:
    %matplotlib notebook
    %matplotlib inline
elif case == 2:
    %matplotlib inline
    from IPython.display import set_matplotlib_formats
    set_matplotlib_formats('pdf')
elif case == 3:
    %matplotlib notebook
font = {'family' : 'times new roman',
        'weight' : 'regular',
        'size'   : 18}
matplotlib.rc('font', **font)
plt.figure()
objects = ('Male', 'Female')
x = np.arange(len(objects))
survive_rate = [maleSR,femaleSR]
 
plt.bar(x, survive_rate, align='center', alpha=0.8)
plt.xticks(x, objects)
plt.ylabel('Survive rate')
# plt.title('Survive rate vs Class')

plt.show()


# Concludion from gender: 
print('female has significantly higher survival rate. This feature is relavent to survival or death')
```

```python
# Pclass
print('Pclass type: ' + str(np.unique(pclass)))
print('total counts: ' + str(len(pclass)))

Pclass1 = (pclass == 1).sum()
Pclass1_survive = ((pclass == 1) & (survive == 1)).sum()
Pclass1_survive_rate = Pclass1_survive/Pclass1
print('Pclass 1 counts: ' + str(Pclass1))
print('Pclass 1 survive counts: ' + str(Pclass1_survive))
print('Pclass 1 survive rate: ' + str(Pclass1_survive_rate))

Pclass2 = (pclass == 2).sum()
Pclass2_survive = ((pclass == 2) & (survive == 1)).sum()
Pclass2_survive_rate = Pclass2_survive/Pclass2
print('Pclass 2 counts: ' + str(Pclass2))
print('Pclass 2 survive counts: ' + str(Pclass2_survive))
print('Pclass 2 survive rate: ' + str(Pclass2_survive_rate))

Pclass3 = (pclass == 3).sum()
Pclass3_survive = ((pclass == 3) & (survive == 1)).sum()
Pclass3_survive_rate = Pclass3_survive/Pclass3
print('Pclass 3 counts: ' + str(Pclass3_survive))
print('Pclass 3 survive counts: ' + str(Pclass3_survive))
print('Pclass 3 survive rate: ' + str(Pclass3_survive_rate))

print('Pclass counts: ' + str(Pclass1 + Pclass2 + Pclass3))
print('\n')
case = 1
if case == 1:
    %matplotlib notebook
    %matplotlib inline
elif case == 2:
    %matplotlib inline
    from IPython.display import set_matplotlib_formats
    set_matplotlib_formats('pdf')
elif case == 3:
    %matplotlib notebook
font = {'family' : 'times new roman',
        'weight' : 'regular',
        'size'   : 18}
matplotlib.rc('font', **font)
plt.figure()
objects = ('PClass 1', 'PClass 2', 'PClass 3')
x = np.arange(len(objects))
survive_rate = [Pclass1_survive_rate,Pclass2_survive_rate,Pclass3_survive_rate]
 
plt.bar(x, survive_rate, align='center', alpha=0.8)
plt.xticks(x, objects)
plt.ylabel('Survive rate')
# plt.title('Survive rate vs Class')

plt.show()
```

```python
# SibSp
sibsp_array = np.unique(sibsp)
sibsp_survive_rate = np.zeros(sibsp_array.max()+1)
for i in range(sibsp_array.max()+1):
    if (sibsp == i).sum() == 0:
        sibsp_survive_rate[i] = 0
    else:
        sibsp_survive_rate[i] = ((sibsp == i) & (survive == 1)).sum()/(sibsp == i).sum()
        
# print('survive rate: ' + str(sibsp_survive_rate))
case = 1
if case == 1:
    %matplotlib notebook
    %matplotlib inline
elif case == 2:
    %matplotlib inline
    from IPython.display import set_matplotlib_formats
    set_matplotlib_formats('pdf')
elif case == 3:
    %matplotlib notebook
font = {'family' : 'times new roman',
        'weight' : 'regular',
        'size'   : 18}
matplotlib.rc('font', **font)
plt.figure()
plt.hist(sibsp,range=[-0.5,8.5],bins = 9)
plt.title('Number of SibSp Distribution')
plt.xlabel('Number of SibSp')
plt.ylabel('Counts')

plt.figure()
# objects = ('PClass 1', 'PClass 2', 'PClass 3')
x = range(sibsp_array.max()+1)
# survive_rate = [Pclass1_survive_rate,Pclass2_survive_rate,Pclass3_survive_rate]

plt.bar(x, sibsp_survive_rate, align='center', alpha=0.8)
# plt.xticks(x, objects)
plt.ylabel('Survive rate')
plt.xlabel('Number of SibSp')

plt.show()
```

```python
# Parch
parch_array = np.unique(parch)
parch_survive_rate = np.zeros(parch_array.max()+1)

for i in range(parch_array.max()+1):
    if (parch == i).sum() == 0:
        parch_survive_rate[i] = 0
    else:
        parch_survive_rate[i] = ((parch == i) & (survive == 1)).sum()/(parch == i).sum()

case = 1
if case == 1:
    %matplotlib notebook
    %matplotlib inline
elif case == 2:
    %matplotlib inline
    from IPython.display import set_matplotlib_formats
    set_matplotlib_formats('pdf')
elif case == 3:
    %matplotlib notebook
font = {'family' : 'times new roman',
        'weight' : 'regular',
        'size'   : 18}
matplotlib.rc('font', **font)
plt.figure()
plt.hist(parch,range=[-0.5,8.5],bins = 9)
plt.title('Number of Parch Distribution')
plt.xlabel('Number of Parch')
plt.ylabel('Counts')

plt.figure()
x = range(parch_array.max()+1)
plt.bar(x, parch_survive_rate, align='center', alpha=0.8)
# plt.xticks(x, objects)
plt.ylabel('Survive rate')
plt.xlabel('Number of Parch')

plt.show()
```
